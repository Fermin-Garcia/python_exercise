```python
import pandas as pd
import numpy as np
```

```python
# Create the users DataFrame.

users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})
users
```

```python
# Create the roles DataFrame

roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
roles
```

```python
# What is the result of using a right join on the DataFrames?
pd.merge(users,roles, how='right', left_on='role_id', right_on = 'id')
```

> we see the variables that do not carry over 

```python
# What is the result of using an outer join on the DataFrames?
pd.merge(users,roles, how='outer', left_on='role_id', right_on = 'id')
```

> we see all the values reguardless of the association of the previously table

```python
# What happens if you drop the foreign keys from the DataFrames and try to merge them?
users.drop(columns='role_id')
pd.merge(users,roles, how='outer', left_on='role_id', right_on = 'id')
```
